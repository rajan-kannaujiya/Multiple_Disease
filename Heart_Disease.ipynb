```python
#import the essential libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
dataset = pd.read_csv('heart.csv')
dataset.head()
```

```python
dataset.shape
```

```python
dataset.describe()
```

```python
dataset['target'].value_counts()
```

```python
# dataset.groupby('target').mean()
```

```python
# seperation of data into dependent and independent dataset.....
#dataset.isnull().sum()
x = dataset.drop(columns='target',axis=1)
y = dataset['target']
x.head()
```

```python
# seperation of the data into training and testing 
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state = 2)
```

```python
print(x.shape,x_train.shape, x_test.shape)
```

```python
# training of the model 
classifier = svm.SVC( kernel= 'linear')
classifier.fit(x_train, y_train)
```

```python
# accuracy of the model 
y_pred = classifier.predict(x_test)
accuracy= accuracy_score(y_pred,y_test )
accuracy
```

```python
prd = classifier.predict([[63,0,1,145,253,0,1,151,1,2.5,2,0,1]])[0]
if( prd==0):
    print("The Person is Having Heart Disease")
else:
    print("The Person is does not have Heart Disease")
```
